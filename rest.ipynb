# GROUPING THE DATA

```python
data={
    'name' : ['anit','virat','rohit','rohan','sayan'],
    'rank' : [1,5,2,3,4],
    'year' : [2005,2004,2005,2001,2004]
}
df=pd.DataFrame(data)
print(df)
res=df.groupby('year').first().reset_index()
print(res[['rank','name','year']])
```

```python
print(df.groupby('year').groups)
```

```python
#mean of group data
import numpy as np
my_data={
    'name' : ['anit','virat','rohit','rohan','sayan'],
    'rank' : [1,5,2,3,4],
    'year' : [2005,2004,2005,2001,2004],
    'salary' : [70000,50000,40000,100000,200]
}
df2=pd.DataFrame(my_data,index=[1,2,3,4,5])
print(df2)
groupres=df2.groupby('year')
print(groupres['salary'].agg(np.mean))
print(groupres['salary'].agg(np.max))
print(groupres['salary'].agg(np.size))
```
